### Geração de tabuleiros

```python
# Mantenha os scripts na mesma pasta do notebook
import slitherlink.gerador as gera
import slitherlink.plota   as plota
```

O algoritmo de geração do tabuleiro é um passeio aleatório + backtracking no grid com objetivo de atingir o ponto de partida. O gerador simplesmente gera tabuleiros de forma aleatória até atingir a densidade mínima especificada (densidade = vértice visitados / tota de vértices do grid) ou atingir o limite de iterações.

Ok, existem algoritmos melhores (esse, por exemplo: https://link.medium.com/JCTiIdeM2zb), mas como não implementava um backtracking desde meu primeiro ano de faculdade fiu para um caminho mais "força-bruta" (e mais caro, e sem garantias de chegar a uma solução).

```python
maxd,tab,cam = gera.gera_Tabuleiro(densidade=0.6
                                   ,lin=30
                                   ,col=30
                                   ,max_tentativas=1000
                                   ,verbose=True
                                   ,seed=1234)
```

```python
plota.plota_tabuleiro(tab,escala=20)
```

Deixei parâmetros para controlar cores, tamanhos e estilos de plotagem. Dá uma olhada na funçao <code>plota.gera_imagem()</code> do módulo plota - essa função gera a imagem do tabuleiro em si. A função <code>plota.plota_tabuleiro()</code> chama e plota usando matplotlib.

```python
help(plota.gera_imagem)
```

Função <code>plota.grava_caminho()</code> gera um vídeo com o caminho do passeio aleatório (ou, se o caminho não for informado, com o caminho do tabuleiro).

```python
plota.grava_caminho(tabuleiro=tab
                     ,arquivo_saida='passeio_aleatorio_exemplo.mp4'
                     ,caminho=cam
                     ,plota_dicas=True
                     ,verbose=True)
```
